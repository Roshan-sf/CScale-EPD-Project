```python
import requests
import os

url = "https://buildingtransparency.org/api/rest-auth/login"

querystring = {"tokenless":"false"}

payload = {
    "username": os.getenv("OPENEPD_USER"),
    "password": os.getenv("OPENEPD_PW")
}
headers = {
    "Content-Type": "application/json",
    "Accept": "application/json"
}

response = requests.post(url, json=payload, headers=headers, params=querystring)

auth_key = response.json()['key']
```

```python
headers = {
    "Accept": "application/json",
    "Authorization": "Bearer "+auth_key
}
```

```python
industry_epd_url = "https://buildingtransparency.org/api/industry_epds"

querystring = {"page_size":"250"}

response = requests.get(industry_epd_url, headers=headers, params=querystring)

response_data = response.json()
```

```python
total_pages = int(response.headers['X-Total-Pages'])
```

```python
industry_epds = []

for i in range(total_pages+1):
        querystring = {
                "page_number": i+1,
                "page_size":"250"}
        response = requests.get(industry_epd_url, headers=headers, params=querystring)
        epd_data = response.json()
        for epd in epd_data:
                industry_epds.append(epd)
        print("successfully completed page "+str(i+1))
        
```

```python
expired_industry_epds = []
valid_industry_epds = []
unknown_industry_epds = []
expiry_date = datetime.strptime("2024-12-31", "%Y-%m-%d")

for epd in industry_epds:
        if type(epd) == dict:
            if epd["date_validity_ends"] is not None:
                validity_date = datetime.strptime(epd["date_validity_ends"], "%Y-%m-%d")
                if validity_date < expiry_date:
                    expired_industry_epds.append(epd)
                else:
                    valid_industry_epds.append(epd)
            else:
                unknown_industry_epds.append(epd)
        else:
            unknown_industry_epds.append(epd)
```

```python
product_epd_ids = []
epd_url = "https://openepd.buildingtransparency.org/api/epds"
headers = {
    "Accept": "application/json",
    "Content-Type": "application/json",
    "Authorization": "Bearer "+auth_key
}

for i in range(5):
        querystring = {
                "page_number": i+1,
                "page_size":"250"}
        response = requests.get(epd_url, headers=headers, params=querystring)
        product_epd_data = response.json()
        for epd in product_epd_data:
                product_epd_ids.append(epd['id'])
        print("successfully completed page "+str(i+1))
```

```python
product_epd_data = []
failed_epd_ids = []
for id in product_epd_ids:
        impact_info_url = "https://openepd.buildingtransparency.org/api/epds/"+str(id)
        try:
                response_impact = requests.get(impact_info_url, headers=headers)
                product_epd_data.append(response_impact.json())
                print("success: "+ id)
        except:
                failed_epd_ids.append(id)
```
